## Web-sraping Top Tech Companies 

```python
#Importing libraries for scraping

from bs4 import BeautifulSoup
import requests
```

```python
#Connecting python to the website for requests 

url = "https://en.wikipedia.org/wiki/List_of_largest_technology_companies_by_revenue"

headers = {"User-Agent": "Chrome/122.0.0.0 Safari/537.36"}

requests.get(url, headers = headers)
```

```python
page = requests.get(url, headers = headers)

print(page.status_code)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
soup.find_all('table', class_='wikitable')
```

```python
tables = soup.find_all('table', class_='wikitable')
```

```python
tables
```

```python
# There is more than one table with the same class on the website so I indexed the particular table we want like it's a list

tables[1]
```

```python
# Stored the table as second_table for continuous identification

second_table = tables[1]
```

```python
second_table
```

```python
# Locating the Table heads for extraction to proceed to cleaning.

second_table.find_all('th')
```

```python

large_companies_titles = second_table.find_all('th')

print(large_companies_titles)
```

```python
print(type(large_companies_titles))
```

```python
# Cleaning the headers

large_companies_titles = [title.text.strip() for title in large_companies_titles]

print(large_companies_titles)
```

```python
#importing pandas to convert the data into dataframe

import pandas as pd 
```

```python
#storing the data as columns in a dataframe 

df = pd.DataFrame(columns = large_companies_titles)

df
```

```python
# Locating the rows of the table 

second_table.find_all('tr')
```

```python
column_frame = second_table.find_all('tr')
```

```python
#Cleaning the rows in the table to be extracted

for row in column_frame[1:]:
    row_frame = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_frame]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```
